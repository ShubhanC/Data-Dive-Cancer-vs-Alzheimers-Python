```python
import pandas as pd
import cancerVsAlzheimers as CA
```

```python
def cleanCancerData(fileName):
    df = pd.read_csv("datasets/" + fileName, skiprows=8)
    df = df.dropna()
    df.index -= 1
    return df
```

```python
df_us = cleanCancerData("incd_us.csv")
df_us = df_us[(df_us["Lower 95% Confidence Interval"] != "data not available") & (df_us["Average Annual Count"] != "3 or fewer")]
df_us = df_us.drop(columns=["Lower CI (CI*Rank)", "CI*Rank([rank note])", "Upper CI (CI*Rank)"])
df_us
```
